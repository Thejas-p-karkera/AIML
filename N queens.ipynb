```python
def nqueens(n, row=0, board=None, sols=None):
    if board is None: board = [-1]*n
    if sols is None: sols = []
    if row == n:
        sols.append(board[:]); return sols
    for col in range(n):
        if all(board[r] != col and abs(board[r]-col) != abs(r-row) for r in range(row)):
            board[row] = col
            nqueens(n, row+1, board, sols)
    return sols

def show(n):
    sols = nqueens(n)
    print("Total:", len(sols))
    for k, sol in enumerate(sols, 1):
        print(f"\nSolution {k}:")
        for r in range(n):
            print(" ".join("Q" if c == sol[r] else "." for c in range(n)))

n = int(input("Enter starting node for BFS: "))

show(n)
```
