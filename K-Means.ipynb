```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\diabetes.csv")

# Scale only the features used for clustering
scaler = StandardScaler()
X = scaler.fit_transform(df[['Glucose', 'BloodPressure']])

# Elbow method
wcss = []
for k in range(1, 18):
    km = KMeans(n_clusters=k, init='k-means++', random_state=0)
    km.fit(X)
    wcss.append(km.inertia_)

plt.plot(range(1, 18), wcss, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.grid(True)
plt.show()

# Final KMeans with chosen k
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
df['cluster'] = kmeans.fit_predict(X)

# Scatter plot
sns.scatterplot(x='Glucose', y='BloodPressure',data=df, hue='cluster')
plt.title("KMeans Clustering (Glucose vs BloodPressure)")
plt.show()

print("Cluster Centers (scaled):")
print(kmeans.cluster_centers_)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\players_22.csv")

# Keep only needed columns and drop rows with NaN in them
features = ['overall', 'potential', 'value_eur', 'wage_eur', 'age']
df_clean = df[features].dropna()          # removes rows with NaN in any of these columns

# Scale features
scaler = StandardScaler()
X = scaler.fit_transform(df_clean)

# Elbow method
wcss = []
for k in range(1, 18):
    km = KMeans(n_clusters=k, init='k-means++', random_state=0)
    km.fit(X)
    wcss.append(km.inertia_)

plt.plot(range(1, 18), wcss, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method (FIFA 22)")
plt.grid(True)
plt.show()

# Final KMeans with chosen k
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
df_clean['cluster'] = kmeans.fit_predict(X)

# Scatter plot
sns.scatterplot(x='overall', y='potential', data=df_clean, hue='cluster', palette='Set1')
plt.title("KMeans Clustering (overall vs potential)")
plt.show()

print("Cluster Centers (scaled):")
print(kmeans.cluster_centers_)
```
