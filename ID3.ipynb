```python
#iris dataset
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")

df.drop(columns=["Id"], inplace=True)

le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])
    
X = df.drop(columns=["Species"])
y = df["Species"]

model = DecisionTreeClassifier(criterion="entropy")
model.fit(X, y)

plt.figure(figsize=(10, 10))
plot_tree(model, feature_names=X.columns, class_names=["Iris-setosa", "Iris-versicolor","Iris-virginica"], filled=True)
plt.title("Decision Tree (ID3 Algorithm)")
plt.show()
```

```python
#titanic dataset ID3
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\titanic.csv")
df.drop(columns=["PassengerId"], inplace=True)

le = LabelEncoder()

for column in df.columns:
    df[column] = le.fit_transform(df[column])

X = df.drop(columns=["Sex"])
y = df["Sex"]

model = DecisionTreeClassifier(criterion="entropy")
model.fit(X, y)

plt.figure(figsize=(10, 10))
plot_tree(model, feature_names=X.columns, class_names=["female", "male"], filled=True)
plt.title("Decision Tree (ID3 Algorithm)")
plt.show()
```
