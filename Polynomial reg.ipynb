# Polynomial (Quadratic regression model for the following data)-using matrix method

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")

x = df["SepalLengthCm"]  # Independent variable
y = df["SepalWidthCm"]          # Dependent variable
n = len(x)

# Step 1: Create the design matrix for quadratic regression
# Each row of X is [1, x_i, x_i^2]
X = np.vstack([np.ones(n), x, x**2]).T

# Step 2: Compute matrices A = XᵀX and B = Xᵀy
A = X.T @ X
B = X.T @ y

# Step 3: Define a function to compute the adjoint of a 3x3 matrix
def adjoint_3x3(M):
    adj = np.zeros((3, 3))
    for i in range(3):
        for j in range(3):
            minor = np.delete(np.delete(M, i, axis=0), j, axis=1)
            cofactor = ((-1) ** (i + j)) * np.linalg.det(minor)
            adj[j, i] = cofactor  # Transpose while assigning
    return adj

# Step 4: Compute the inverse of A using its adjoint and determinant
det_A = np.linalg.det(A)
adj_A = adjoint_3x3(A)
A_inv = adj_A / det_A

# Step 5: Solve for regression coefficients: alpha = A⁻¹ * B
alpha = A_inv @ B
a0, a1, a2 = alpha

#Print the full equation
print("y=", a2 ,"x² + ",a1 ,"x + ",a0)
```

# Polynomial (Quadratic regression model for the following data)-using matrix method using library function

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")

x = df["SepalLengthCm"]  # Independent variable
y = df["SepalWidthCm"]          # Dependent variable
n = len(x)

# Fit a 2nd-degree polynomial: y = a2*x^2 + a1*x + a0
coeffs = np.polyfit(x, y, deg=2)  # Returns [a2, a1, a0]
a2, a1, a0 = coeffs

# Print the full formula without using f-string
print("y =", a2, "x² +", a1, "x +", a0)
```
