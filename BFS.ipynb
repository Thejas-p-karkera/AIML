```python
from collections import defaultdict, deque
import networkx as nx
import matplotlib.pyplot as plt

graph = defaultdict(list)

# Number of edges
edges = int(input("Enter number of edges: "))

print("Enter edges (u v):")
for _ in range(edges):
    u, v = map(int, input().split())
    graph[u].append(v)

# BFS function
def bfs(start):
    visited = set()
    queue = deque([start])
    visited.add(start)

    order = []

    while queue:
        node = queue.popleft()
        order.append(node)

        for nbr in graph[node]:
            if nbr not in visited:
                visited.add(nbr)
                queue.append(nbr)

    return order

# Starting node
start = int(input("Enter starting node for BFS: "))
order = bfs(start)

print("BFS Traversal:")
print(*order)

# -------- GRAPH DRAWING -------- #

G = nx.Graph()   # Undirected (matches your output)

for u in graph:
    for v in graph[u]:
        G.add_edge(u, v)

plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True)

plt.show()
```
