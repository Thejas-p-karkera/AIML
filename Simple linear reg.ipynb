# Simple linear Regression(Using Library function)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Sample data
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")

x = df[["SepalLengthCm"]]  # Independent variable
y = df["SepalWidthCm"]          # Dependent variable

# Create and train the model
model = LinearRegression()
model.fit(x, y)

# Predict values
y_pred = model.predict(x)

# Print model parameters
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

# Calculate error metrics
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)

# Plot actual vs predicted
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

# simple linear function without Library function

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")

x = df["SepalLengthCm"]  # Independent variable
y = df["SepalWidthCm"]

# Step 1: Calculate slope (m) and intercept (b) manually
#n = len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate slope (m)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
slope = numerator / denominator

# Calculate intercept (b)
intercept = y_mean - slope * x_mean

print("Slope (Coefficient):", slope)
print("Intercept:", intercept)

# Step 2: Predict values
y_pred = slope * x + intercept

# Step 3: Calculate error metrics manually
mse = np.mean((y - y_pred) ** 2)
mae = np.mean(np.abs(y - y_pred))
rmse = np.sqrt(mse)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)

# Step 4: Plot actual vs predicted
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Manual Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

# Simple Linear Regression(using Matrix method)

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")

x = df["SepalLengthCm"]  # Independent variable
y = df["SepalWidthCm"]

# Step 1: Prepare the design matrix X and target vector Y
X = np.vstack([np.ones(len(x)), x]).T  # Add intercept term
Y = y.values.reshape(-1, 1)

# Step 2: Compute coefficients using the Normal Equation:  = (XX)XY
theta = np.linalg.inv(X.T @ X) @ X.T @ Y
intercept, slope = theta.flatten()

print("Slope (Coefficient):", slope)
print("Intercept:", intercept)

# Step 3: Predict values
y_pred = X @ theta
y_pred = y_pred.flatten()

# Step 4: Calculate error metrics
mse = np.mean((y - y_pred) ** 2)
mae = np.mean(np.abs(y - y_pred))
rmse = np.sqrt(mse)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)

# Step 5: Plot actual vs predicted
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression (Matrix Method)')
plt.legend()
plt.grid(True)
plt.show()
```
