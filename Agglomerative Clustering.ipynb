# Single Linkage

```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Load Iris data
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

# Single linkage agglomerative clustering
Z = linkage(X, method='single')

# FIXED: Use sample index labels (0 to 149)
plt.figure(figsize=(12, 6))
dendrogram(Z, labels=range(len(X)))
plt.title('Single Linkage Agglomerative Clustering - Iris Dataset')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

# Complete Linkage

```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt

# Load Iris data
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

# Complete linkage (MAXIMUM distance between clusters)
Z = linkage(X, method='complete')

# Dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z, labels=range(len(X)))
plt.title('Complete Linkage Agglomerative Clustering - Iris Dataset')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

# Average Linkage

```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt

# Load Iris data
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

# Average linkage (MEAN distance between all pairs in clusters)
Z = linkage(X, method='average')

# Dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z, labels=range(len(X)))
plt.title('Average Linkage Agglomerative Clustering - Iris Dataset')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```
