```python
#iris dataset
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load and preprocess the dataset
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\Iris.csv")
df.drop(columns=["Id"], inplace=True)

# Encode categorical columns
le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])

# Split features and target
X = df.drop(columns=["Species"])
y = df["Species"]

# Train decision tree using entropy (C4.5-style
model = DecisionTreeClassifier(criterion="entropy")
model.fit(X, y)

# Visualize the tree
plt.figure(figsize=(10, 8))
plot_tree(model, feature_names=X.columns, class_names=["Iris-setosa", "Iris-versicolor", "Iris-virginica"], filled=True)
plt.title("Decision Tree (C4.5 Style - Gain Ratio Approximation)")
plt.show()
```

```python
#titanic dataset c4.5
#iris dataset
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load and preprocess the dataset
df = pd.read_csv(r"D:\Coding\MSc\3rd sem\Datasets\AI&ML_Dataset\titanic.csv")
df.drop(columns=["PassengerId"], inplace=True)

# Encode categorical columns
le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])

# Split features and target
X = df.drop(columns=["Sex"])
y = df["Sex"]

# Train decision tree using entropy (C4.5-style
model = DecisionTreeClassifier(criterion="entropy")
model.fit(X, y)

# Visualize the tree
plt.figure(figsize=(10, 8))
plot_tree(model, feature_names=X.columns, class_names=["female", "male"], filled=True)
plt.title("Decision Tree (C4.5 Style - Gain Ratio Approximation)")
plt.show()
```
