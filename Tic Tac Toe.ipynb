```python
import tkinter as tk
from tkinter import messagebox

P,A="X","O"
W=[(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
root=tk.Tk()
B=[]

def win(p):
    return any(B[i]["text"]==B[j]["text"]==B[k]["text"]==p for i,j,k in W)

def reset():
    for b in B: b["text"]=""

def end():
    if win(P): messagebox.showinfo("Result","You win"); reset(); return 1
    if win(A): messagebox.showinfo("Result","AI wins"); reset(); return 1
    if all(b["text"] for b in B): messagebox.showinfo("Result","Draw"); reset(); return 1

def ai():
    for p in (A,P):
        for i in range(9):
            if not B[i]["text"]:
                B[i]["text"]=p
                if win(p): B[i]["text"]=A; end(); return
                B[i]["text"]=""
    for i in [4,0,2,6,8,1,3,5,7]:
        if not B[i]["text"]:
            B[i]["text"]=A; end(); return

def click(i):
    if not B[i]["text"]:
        B[i]["text"]=P
        if not end(): root.after(150,ai)

for i in range(9):
    b=tk.Button(root,font=("Arial",24),width=4,height=2,
                command=lambda i=i:click(i))
    b.grid(row=i//3,column=i%3); B.append(b)

root.mainloop()
```
