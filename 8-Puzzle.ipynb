```python
from collections import deque

goal = (1,2,3,4,5,6,7,8,0)

def print_board(s):
    for i in range(0,9,3):
        print(" ".join(str(x) if x else "_" for x in s[i:i+3]))
    print()

def solve(start):
    q = deque([start])
    parent = {start: None}

    while q:
        s = q.popleft()

        if s == goal:
            path = []
            while s:
                path.append(s)
                s = parent[s]
            path.reverse()

            print("Start:")
            print_board(path[0])

            print("Goal:")
            print_board(goal)

            print("Solution in", len(path)-1, "moves\n")
            for i, state in enumerate(path):
                print("Step", i)
                print_board(state)
            return

        i = s.index(0)
        for j in [i-3, i+3, i-1, i+1]:
            if 0 <= j < 9:
                t = list(s)
                t[i], t[j] = t[j], t[i]
                t = tuple(t)
                if t not in parent:
                    parent[t] = s
                    q.append(t)

solve((1,2,3,4,0,6,7,5,8))
```
