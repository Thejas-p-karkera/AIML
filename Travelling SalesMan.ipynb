```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.approximation import traveling_salesman_problem

# Graph
G = nx.Graph()
G.add_weighted_edges_from([
    ('A','B',10), ('A','C',15), ('A','D',20),
    ('B','C',35), ('B','D',25),
    ('C','D',30)
])

# TSP
tour = traveling_salesman_problem(G, weight='weight', cycle=True)
cost = sum(G[u][v]['weight'] for u, v in zip(tour[:-1], tour[1:]))

print("Tour:", tour)
print("Cost:", cost)

# Fixed positions
pos = nx.spring_layout(G)

# Draw graph
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos,
    edge_labels=nx.get_edge_attributes(G,'weight'))

# Highlight best path (thicker edges only)
best_edges = list(zip(tour[:-1], tour[1:]))
nx.draw_networkx_edges(G, pos, edgelist=best_edges, width=3)

plt.show()
```
